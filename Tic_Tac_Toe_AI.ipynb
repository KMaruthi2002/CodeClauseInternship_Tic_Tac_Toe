```python
class TicTacToe:
    def __init__(self):
        self.board = [' ' for _ in range(9)]  # A list to hold the board state
        self.current_player = 'X'  # Player X starts
```

```python
def display_board(self):
        print(f"{self.board[0]} | {self.board[1]} | {self.board[2]}")
        print("--+---+--")
        print(f"{self.board[3]} | {self.board[4]} | {self.board[5]}")
        print("--+---+--")
        print(f"{self.board[6]} | {self.board[7]} | {self.board[8]}")
        def check_winner(self):
        winning_combinations = [
            (0, 1, 2), (3, 4, 5), (6, 7, 8),  # Horizontal
            (0, 3, 6), (1, 4, 7), (2, 5, 8),  # Vertical
            (0, 4, 8), (2, 4, 6)               # Diagonal
        ]
        for combo in winning_combinations:
            if self.board[combo[0]] == self.board[combo[1]] == self.board[combo[2]] != ' ':
                return self.board[combo[0]]
        return None

    def is_draw(self):
        return ' ' not in self.board

    def make_move(self, position):
        if self.board[position] == ' ':
            self.board[position] = self.current_player
            return True
        return False
```

```python
def switch_player(self):
        self.current_player = 'O' if self.current_player == 'X' else 'X'

    def minimax(self, is_maximizing):
        winner = self.check_winner()
        if winner == 'X':
            return -1  # Player X (you) loses
        elif winner == 'O':
            return 1  # Player O (AI) wins
        elif self.is_draw():
            return 0  # Draw

        if is_maximizing:
            best_score = -float('inf')
            for i in range(9):
                if self.board[i] == ' ':
                    self.board[i] = 'O'  # AI's move
                    score = self.minimax(False)
                    self.board[i] = ' '  # Undo move
                    best_score = max(score, best_score)
            return best_score
        else:
            best_score = float('inf')
            for i in range(9):
                if self.board[i] == ' ':
                    self.board[i] = 'X'  # Player's move
                    score = self.minimax(True)
                    self.board[i] = ' '  # Undo move
                    best_score = min(score, best_score)
            return best_score
```

```python
 def ai_move(self):
        best_score = -float('inf')
        best_move = None
        for i in range(9):
            if self.board[i] == ' ':
                self.board[i] = 'O'  # AI's move
                score = self.minimax(False)
                self.board[i] = ' '  # Undo move
                if score > best_score:
                    best_score = score
                    best_move = i
        self.make_move(best_move)

    def play_game(self):
        while True:
            self.display_board()
            if self.current_player == 'X':
                position = int(input(f"Player {self.current_player}, enter your move (0-8): "))
                if not self.make_move(position):
                    print("Invalid move. Try again.")
                    continue
            else:
                print(f"AI ({self.current_player}) is making a move...")
                self.ai_move()

            if self.check_winner():
                self.display_board()
                print(f"Player {self.current_player} wins!")
                break
            if self.is_draw():
                self.display_board()
                print("It's a draw!")
                break

            self.switch_player()

if __name__ == "__main__":
    game = TicTacToe()
    game.play_game()
```
